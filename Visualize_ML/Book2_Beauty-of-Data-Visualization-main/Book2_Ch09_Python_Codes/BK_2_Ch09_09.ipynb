```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
# 如果文件夹不存在，创建文件夹
if not os.path.isdir("Figures"):
    os.makedirs("Figures")
```

```python
cmap = mpl.cm.get_cmap('RdYlBu')
num = 1000

start_array = np.random.uniform(0, 3, num)
len_array = np.random.uniform(1, 4, num)
radian_array = np.random.uniform(0, 2*np.pi, num)
color_array = np.random.uniform(0, 1, num)

x_start = np.cos(radian_array) * start_array
y_start = np.sin(radian_array) * start_array

x_end = np.cos(radian_array) * (start_array + len_array)
y_end = np.sin(radian_array) * (start_array + len_array)

fig, ax = plt.subplots()

for x_1_i , y_1_i , x_2_i , y_2_i, c_i in zip(x_start, y_start, 
                                       x_end, y_end, 
                                       color_array):

    ax.plot((x_1_i, x_2_i),
             (y_1_i, y_2_i), 
             color = cmap(c_i), lw = 0.2)

ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
ax.set_aspect('equal')
fig.savefig('Figures/光线.svg', format='svg')
plt.show() 
```

```python
start_array = np.random.uniform(0, 3, num)
len_array = np.random.uniform(1, 4, num)
radian_array = np.random.uniform(0, 2*np.pi, num)

margin = np.pi/360
x_A = np.cos(radian_array - margin) * start_array
y_A = np.sin(radian_array - margin) * start_array

x_B = np.cos(radian_array + margin) * start_array
y_B = np.sin(radian_array + margin) * start_array

x_C = np.cos(radian_array + margin) * (start_array + len_array)
y_C = np.sin(radian_array + margin) * (start_array + len_array)

x_D = np.cos(radian_array - margin) * (start_array + len_array)
y_D = np.sin(radian_array - margin) * (start_array + len_array)

fig, ax = plt.subplots()

for x_A_i , y_A_i , x_B_i , y_B_i , x_C_i , y_C_i , x_D_i , y_D_i, c_i in zip(x_A, y_A, 
                                       x_B, y_B, 
                                       x_C, y_C,
                                       x_D, y_D,
                                       start_array):

    ax.fill((x_A_i, x_B_i, x_C_i, x_D_i, x_A_i),
             (y_A_i, y_B_i, y_C_i, y_D_i, y_A_i), 
             color = cmap(c_i), edgecolor = None, alpha = 0.5)

ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
ax.set_aspect('equal')
fig.savefig('Figures/光线，填充.svg', format='svg')
plt.show()    
```
