# 模数乘法表
Book_2《可视之美》 | 鸢尾花书：从加减乘除到机器学习  

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def mod_multiplication_table(mod):
    i_array = np.arange(mod)
    j_array = np.arange(mod)
    ii,jj = np.meshgrid(i_array,j_array)
    mod_table = ii*jj % mod
    # mod_table = [[(i * j) % mod for j in range(mod)] for i in range(mod)]
    return mod_table

def visualize(mod, annot = True, linecolor = 'k'):
    table_data = mod_multiplication_table(mod)
    table_df = pd.DataFrame(table_data)

    plt.figure(figsize=(8, 6))
    sns.heatmap(table_df, 
                annot=annot, 
                cmap="YlGnBu", 
                fmt="d", 
                linewidths=.05,
                linecolor = linecolor,
                square = True)
    title = f'Mod {mod} Multiplication Table'
    plt.title(title)
    plt.savefig(title + '.svg')
    plt.show()
```

```python
# 以模数5为例
visualize(5)
```

```python
# 以模数36为例
visualize(36, True, None)
```

```python
# 以模数36为例
visualize(36, False, None)
```

```python
# 以模数90为例
visualize(90, False, None)
```

```python
# 以模数180为例
visualize(180, False, None)
```
