$${\displaystyle {\begin{aligned}x(\varphi )&=2a(1-\cos \varphi )\cdot \cos \varphi \ ,\\y(\varphi )&=2a(1-\cos \varphi )\cdot \sin \varphi \ ,\qquad 0\leq \varphi <2\pi \end{aligned}}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = plt.rcParams
p["font.sans-serif"] = ["Roboto"]
p["font.weight"] = "light"
p["ytick.minor.visible"] = True
p["xtick.minor.visible"] = True
p["axes.grid"] = True
p["grid.color"] = "0.5"
p["grid.linewidth"] = 0.5
```

```python
a = 1
phi_array = np.linspace(0,2*np.pi,181)
x_array = 2*a*(1 - np.cos(phi_array)) * np.cos(phi_array)
y_array = 2*a*(1 - np.cos(phi_array)) * np.sin(phi_array)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x_array,y_array)
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_xticks(np.arange(-5,6))
ax.set_yticks(np.arange(-5,6))
ax.axvline(x = 0, c = 'k')
ax.axhline(y = 0, c = 'k')
ax.grid(True)
# ax.axis('off')
fig.savefig('心形线，参数方程，1.svg') # png
```
