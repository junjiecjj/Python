# 利用melt() 将宽格式转换为长格式
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习  

```python
import pandas as pd
```

```python
data = {
    'Student ID': ['1', '2', '3', '4'],
    'Art':        [4, 3, 5, 4],
    'Math':       [3, 4, 5, 3],
    'Science':    [5, 4, 3, 4]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.columns.names = ['Subject']
df
```

```python
melted_df = df.melt(id_vars='Student ID', 
                    var_name='Subject', 
                    value_vars=['Art','Math','Science'], 
                    value_name='Score')
melted_df
```
