# 利用unstack() 将长格式转换为宽格式
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习  

```python
import pandas as pd
import numpy as np
```

```python
# 创建班级、学号和科目的所有可能组合
classes =  ['A', 'B']
student_ids = [1, 2, 3, 4]
subjects =    ['Art', 'Math', 'Science']
```

```python
# 使用随机数生成成绩数据
length = len(classes)*len(student_ids)*len(subjects)
scores = np.random.randint(3, 6, size=(length))
```

```python
# 创建多级索引
index = pd.MultiIndex.from_product(
    [classes, student_ids, subjects], 
    names=['Class', 'Student ID', 'Subject'])
```

```python
# 创建数据帧
df = pd.DataFrame(scores, index=index, 
    columns=['Final'])
df
```

```python
# df.unstack(0)
df.unstack('Class')
```
