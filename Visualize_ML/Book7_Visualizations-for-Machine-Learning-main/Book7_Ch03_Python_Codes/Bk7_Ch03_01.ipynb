```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import yfinance as yf
```

```python
p = plt.rcParams
p["font.sans-serif"] = ["Roboto"]
p["font.weight"] = "light"
p["ytick.minor.visible"] = False
p["xtick.minor.visible"] = False
p["axes.grid"] = True
p["grid.color"] = "0.5"
p["grid.linewidth"] = 0.5
```

```python
y_X_df = yf.download(['AAPL','MCD','^GSPC'], 
                     start='2020-01-01', 
                     end='2020-12-31')

y_X_df.to_pickle('y_X_df.pkl')
y_X_df = y_X_df['Adj Close'].pct_change()
y_X_df.dropna(inplace = True)
```

```python
y_X_df.rename(columns={"^GSPC": "SP500"},
              inplace = True)

X_df = y_X_df[['AAPL','MCD']]
y_df = y_X_df[['SP500']]
```

## Data analysis

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(X_df["AAPL"], X_df["MCD"], y_df,
           s = 8, alpha = 0.5)

ax.set_xlabel('AAPL')
ax.set_ylabel('MCD')
ax.set_zlabel('SP500')
ax.set_proj_type('ortho')
ax.set_xlim([-0.15,0.15])
ax.set_ylim([-0.15,0.15])
ax.set_zlim([-0.15,0.15])
plt.savefig('三维散点数据.svg')
```

```python
g = sns.pairplot(y_X_df)
g.map_upper(sns.scatterplot, color = 'b')
g.map_lower(sns.kdeplot, levels=8, fill=True, cmap="viridis_r") 
g.map_diag(sns.histplot, kde=False, color = 'b')
```

## covariance matrix

```python
SIGMA = y_X_df.cov()
```

```python
fig, axs = plt.subplots()

h = sns.heatmap(SIGMA, annot=True,cmap='RdBu_r')
h.set_aspect("equal")
```

## correlation matrix

```python
RHO = y_X_df.corr()

fig, axs = plt.subplots()

h = sns.heatmap(RHO, annot=True,cmap='RdBu_r')
h.set_aspect("equal")
```

## Volatility vector space

```python
Angles = np.arccos(RHO)*180/np.pi
fig, axs = plt.subplots()

h = sns.heatmap(Angles, annot=True,cmap='RdBu_r')
h.set_aspect("equal")
```

## Regression

```python
# add a column of ones
X_df = sm.add_constant(X_df)
```

```python
model = sm.OLS(y_df, X_df)
results = model.fit()
print(results.summary())
```

```python
p = model.fit().params
print(p)
```

```python
# generate x-values for your regression line (two is sufficient)
xx1,xx2 = np.meshgrid(np.linspace(-0.15,0.15,20), 
                      np.linspace(-0.15,0.15,20))

yy = p.AAPL*xx1 + p.MCD*xx2 + p.const
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(X_df["AAPL"], X_df["MCD"], y_df,
           s = 8, alpha = 0.5)
ax.plot_wireframe(xx1, xx2, yy, color = 'r')
ax.set_xlim([-0.15,0.15])
ax.set_ylim([-0.15,0.15])
ax.set_zlim([-0.15,0.15])
ax.set_xlabel('AAPL')
ax.set_ylabel('MCD')
ax.set_zlabel('SP500')
ax.set_proj_type('ortho')
plt.savefig('三维散点数据 + 回归平面.svg')
```
