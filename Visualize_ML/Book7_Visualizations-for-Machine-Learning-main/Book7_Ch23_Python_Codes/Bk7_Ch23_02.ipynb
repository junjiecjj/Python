```python
from sklearn.metrics.pairwise import rbf_kernel
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import datasets
```

```python
p = plt.rcParams
p["font.sans-serif"] = ["Roboto"]
p["font.weight"] = "light"
p["ytick.minor.visible"] = True
p["xtick.minor.visible"] = True
p["axes.grid"] = True
p["grid.color"] = "0.5"
p["grid.linewidth"] = 0.5
```

```python
# load iris data
iris = datasets.load_iris()
X = iris.data[:, [0,1]]

# generate pairwise RBF affinity matrix
rbf_X = rbf_kernel(X)
```

```python
# heatmap of RBF affinity matrix
fig, ax = plt.subplots()

sns.heatmap(rbf_X, cmap="RdYlBu_r",
            square=True)
```

```python
# lower triangle for heatmap of RBF affinity matrix
fig, ax = plt.subplots()

mask = np.triu(np.ones_like(rbf_X, dtype=bool))

sns.heatmap(rbf_X, cmap="RdYlBu_r",
            mask = mask,
            square=True)
```

```python
# Cluster map based on affinity matrix

# fig, ax = plt.subplots()

g = sns.clustermap(rbf_X, cmap="RdYlBu_r")
g.ax_row_dendrogram.remove()
```
