# 二叉树
Book_6《数据有道》 | 鸢尾花书：从加减乘除到机器学习

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.binomial_tree(8)
# n, int
# Order of the binomial tree.

plt.figure(figsize=(8, 8))
nx.draw_networkx(G, node_size=20, 
                 alpha=0.5, 
                 node_color="blue", 
                 with_labels=False)
plt.axis("equal")
plt.savefig('二叉树.svg')
```
