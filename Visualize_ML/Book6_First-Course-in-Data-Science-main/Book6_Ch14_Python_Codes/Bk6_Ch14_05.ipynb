```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 创建完全多分图
G = nx.complete_multipartite_graph(3, 6, 9)
```

```python
# 多分图布局
pos = nx.multipartite_layout(G)
```

```python
# 可视化
plt.figure(figsize = (6,6))
nx.draw_networkx(G, pos = pos, 
                 width = 0.25, 
                 with_labels = False)
plt.savefig('完全多分图.svg')
```
