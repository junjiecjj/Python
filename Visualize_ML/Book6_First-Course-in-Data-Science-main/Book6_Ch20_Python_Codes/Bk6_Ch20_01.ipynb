# 再看邻接矩阵
Book_6《数据有道》 | 鸢尾花书：从加减乘除到机器学习

```python
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
```

```python
G = nx.Graph()
# 创建无向图的实例
```

```python
G.add_nodes_from(['a', 'b', 'c', 'd', 'e', 'f'])
# 添加多个顶点
```

```python
G.add_edges_from([('a','b'),('a','f'),
                  ('a','c'),('a','d'),
                  ('d','e'),('c','e'),
                  ('d','c'),('b','c'),
                  ('b','e'),('b','f'),
                  ('e','f'),('c','f')])

# 增加一组边
```

```python
# 城市具体位置

pos = {'a':(50.6463,6.5692),
       'b':(70.3019,41.0777),
       'c':(35.6363,36.0673),
       'd':(6.8205,16.8171),
       'e':(26.1044,68.8277),
       'f':(59.1871,69.6398)}
```

```python
plt.figure(figsize = (6,6))
nx.draw_networkx(G, 
                 pos = pos,
                 node_size = 180)
plt.savefig('城市位置.svg')
```

```python
A = nx.adjacency_matrix(G).todense()
# 无向图的邻接矩阵
```

```python
A
```

```python
sns.heatmap(A,square = True, 
            cmap = 'Blues',
            annot = True, fmt = '.0f',
            xticklabels = list(G.nodes), 
            yticklabels = list(G.nodes),
            linecolor = 'k', linewidths = 0.2)
plt.savefig('无向图的邻接矩阵A.svg')
```

```python
A @ A
```

```python
sns.heatmap(A @ A,square = True, 
            cmap = 'Blues',
            annot = True, fmt = '.0f',
            xticklabels = list(G.nodes), 
            yticklabels = list(G.nodes),
            linecolor = 'k', linewidths = 0.2)
plt.savefig('A @ A.svg')
```

```python
A @ A @ A
```

```python
sns.heatmap(A @ A @ A,square = True, cmap = 'Blues', annot = True)
```

```python
A + A @ A + A @ A @ A
```

```python
sns.heatmap(A + A @ A + A @ A @ A,square = True, cmap = 'Blues', annot = True)
```
