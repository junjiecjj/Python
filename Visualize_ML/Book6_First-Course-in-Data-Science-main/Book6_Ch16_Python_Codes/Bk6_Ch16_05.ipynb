# 桥、局部桥
Book_6《数据有道》 | 鸢尾花书：从加减乘除到机器学习

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python

# G = nx.balanced_tree(3, 5)
G = nx.karate_club_graph()
# 创建随机图
pos = nx.spring_layout(G, seed = 2)
```

```python
plt.figure(figsize = (6,6))

nx.draw_networkx(G, pos, 
                 node_size = 18,
                 with_labels = False)
plt.savefig('全图.svg')
```

```python
nx.has_bridges(G)
```

```python
#%%
bridges = list(nx.bridges(G))
len(bridges)
```

```python
plt.figure(figsize = (6,6))

nx.draw_networkx(G, pos,
                 with_labels = False)

nx.draw_networkx_nodes(G,
                       pos=pos, 
                       nodelist = set(sum(bridges, ())),
                       node_color = 'r')

nx.draw_networkx_edges(
    G, pos, edgelist=bridges, width=1, edge_color="r"
)  # red color for bridges

# plt.axis("off")
plt.savefig('桥.svg')
```

```python
#%% 局部桥


local_bridges = list(nx.local_bridges(G, with_span=False))
len(local_bridges)
```

```python
plt.figure(figsize = (6,6))

nx.draw_networkx(G, pos,
                 with_labels = False)

nx.draw_networkx_nodes(G,
                       pos=pos, 
                       nodelist = set(sum(local_bridges, ())),
                       node_color = 'pink')

nx.draw_networkx_edges(G, pos, 
                       edgelist=local_bridges, 
                       width=1, edge_color="pink")

# plt.axis("off")
plt.savefig('局部桥.svg')
```

```python
plt.figure(figsize = (6,6))

nx.draw_networkx(G, pos,
                 with_labels = False)

nx.draw_networkx_edges(G, pos, 
                       edgelist=set(local_bridges) - set(bridges), 
                       width=1, edge_color="orange")  

nx.draw_networkx_nodes(G,
                       pos=pos, 
                       nodelist = set(sum(list(set(local_bridges) - set(bridges)), ())),
                       node_color = 'orange')

# plt.axis("off")
plt.savefig('局部桥中不是桥的边.svg')
```
