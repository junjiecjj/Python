# 绘制十二面体图，图着色问题
Book_6《数据有道》 | 鸢尾花书：从加减乘除到机器学习

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 创建十二面体图
G = nx.dodecahedral_graph()
```

```python
# 贪心着色算法
graph_color_code = nx.greedy_color(G)
```

```python
graph_color_code
```

```python
# 独特颜色，{0, 1, 2, 3}
unique_colors = set(graph_color_code.values())
```

```python
unique_colors
```

```python
# 颜色映射
color_mapping = {0: '#0099FF', 
                 1: '#FF6600', 
                 2: '#99FF33', 
                 3: '#FF99FF'}
```

```python
# 完成每个节点的颜色映射
node_colors = [color_mapping[graph_color_code[n]] 
               for n in G.nodes()]

# 节点位置布置
pos = nx.spring_layout(G, seed=14)

# 可视化
fig, ax = plt.subplots(figsize = (6,6))
nx.draw_networkx(
    G,
    pos,
    with_labels=True,
    node_size=500,
    node_color=node_colors,
    edge_color="grey",
    font_size=12,
    font_color="#333333",
    width=2,
)
plt.savefig('十二面体图，着色问题.svg')
```
