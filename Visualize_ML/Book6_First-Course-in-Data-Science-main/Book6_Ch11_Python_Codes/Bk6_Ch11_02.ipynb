```python
import networkx as nx
import matplotlib.pyplot as plt
```

## 第一幅图

```python
G = nx.cubical_graph()
# 立方体图
```

```python
plt.figure(figsize = (6,6))

nx.draw_networkx(G, 
                 pos = nx.spring_layout(G, seed = 8),
                 with_labels=True, 
                 node_color="c")
plt.savefig('图G.svg')
```

## 第二幅图

```python
H = nx.Graph()
H.add_edges_from([('a','b'),('b','c'), 
                  ('c','d'),('e','f'),
                  ('f','g'),('g','h'),
                  ('b','h'),('c','g'),
                  ('d','a'),('e','h'),
                  ('e','a'),('d','f')])
```

```python
plt.figure(figsize = (6,6))

nx.draw_networkx(H, 
                 pos = nx.circular_layout(H),
                 with_labels=True, 
                 node_color="orange")
plt.savefig('图H.svg')
```

```python
nx.is_isomorphic(G,H)
# 判断是否同构
```

```python
nx.vf2pp_isomorphism(G,H, node_label="label")
# 节点对应关系
```
