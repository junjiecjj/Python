# 节点间路径，无向图
Book_6《数据有道》 | 鸢尾花书：从加减乘除到机器学习

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.complete_graph(5)
# 完全图
```

```python
# 可视化图
plt.figure(figsize = (6,6))
pos = nx.circular_layout(G)
nx.draw_networkx(G, 
                 pos = pos,
                 with_labels = True,
                 node_size = 188) 
plt.savefig('完全图.svg')
```

```python
# 节点0、3之间所有路径

all_paths_nodes = nx.all_simple_paths(G, source=0, target=3)
# 节点0、3之间所有路径上的节点

all_paths_edges = nx.all_simple_edge_paths(G, source=0, target=3)
# 节点0、3之间所有路径上的边

fig, axes = plt.subplots(4, 4, 
                         figsize = (8,8))

axes = axes.flatten()

for nodes_i, edges_i, ax_i in zip(all_paths_nodes,
                                  all_paths_edges,
                                  axes):
    
    nx.draw_networkx(G, 
                     ax = ax_i,
                     pos = pos,
                     with_labels = True,
                     node_size = 88) 
    
    nx.draw_networkx_nodes(G,
                           ax = ax_i,
                           nodelist = nodes_i,
                           pos = pos,
                           node_size = 88,
                           node_color = 'r')

    nx.draw_networkx_edges(G,
                           pos = pos,
                           ax = ax_i,
                           edgelist = edges_i,
                           edge_color = 'r')
        
    ax_i.set_title(' → '.join(str(node) for node in nodes_i))
    ax_i.axis('off')
    
plt.savefig('节点0、3之间所有路径.svg')   
```

```python
# 最短路径
print(nx.shortest_path(G, source=0, target=3))
```
