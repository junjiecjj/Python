## 更复杂的图

```python
seed = 89
# 创建图
G = nx.gnp_random_graph(100, 0.5, seed=seed)
pos = nx.spring_layout(G, seed = 88)
```

```python
# 特征向量中心性
centrality = nx.eigenvector_centrality(G)
```

```python
# 取出特征向量中心性具体值
list_c = np.array(list(centrality.values()))
```

```python
plt.hist(list_c,ec = 'k')
plt.ylabel('Count')
plt.xlabel('Eigenvector centrality')
plt.savefig('特征向量中心性分布.svg')
```

```python
max_c = max(list_c)
min_c = min(list_c)
```

```python
# 以特征向量中心性大小设定节点大小
node_size = 100 * (list_c - min_c)/(max_c - min_c)
```

```python
# 可视化
fig, ax = plt.subplots(figsize=(8, 8))
nx.draw_networkx(G, pos=pos, 
                 node_size=node_size, 
                 edge_color = '0.18',
                 with_labels=False, 
                 cmap = 'RdYlBu_r',
                 node_color = node_size,
                 alpha = 0.68,
                 width=0.15)
plt.axis("off")
plt.savefig('特征向量中心性.svg')
```
