[**第九章：非线性滤波**](./09-非线性滤波.ipynb)

卡尔曼滤波器仅适用于线性问题，然而这个世界是非线性的。在这里，我将介绍非线性系统对滤波器带来的挑战，并简要讨论我们将在后面几章学习的各种算法。


[**第十章：无迹卡尔曼滤波器**](./10-无迹卡尔曼滤波器.ipynb)

无迹卡尔曼滤波器（UKF）是卡尔曼滤波理论的最新发展，它允许你过滤非线性问题，而不像扩展卡尔曼滤波器那样需要一个封闭式的解决方案。

现有资料中大多数讨论都是关于扩展卡尔曼滤波器，而这个主题很少被提及，我把它放到前面是因为UKF的理解和实现都要容易得多，并且同其滤波器性能通常等同甚至要优于扩展卡尔曼滤波器。对于现实世界中的问题我通常首先尝试使用UKF来解决，建议你也这么做。


[**第十一章：扩展卡尔曼滤波器**](./11-扩展卡尔曼滤波器.ipynb)

扩展卡尔曼滤波器（EKF）是非线性问题线性化最常用的方法。大多数现实中的卡尔曼滤波器都是EKF，因此需要了解这个以方便理解现有的一些代码、论文和讨论等。


[**第十二章：粒子滤波器**](./12-粒子滤波器.ipynb)

粒子滤波器使用蒙特卡罗技术来对数据进行滤波。它们很容易处理强非线性和非高斯系统，以及多模分布（同时跟踪多个对象），代价是高计算量需求。


[**第十三章：平滑化**](./13-平滑化.ipynb)

卡尔曼滤波器是递归的，因此非常适合于实时滤波。但是，它们在后处理数据方面工作得非常好。毕竟，卡尔曼滤波器是预测校正器，预测过去比预测未来更容易！我们讨论一些常见的方法。


[**第十四章：自适应滤波**](./14-自适应滤波.ipynb)
 
卡尔曼滤波器假设了一个单一的过程模型，但通常需要用几个不同的过程模型来描述制造目标。自适应滤波利用多种技术使卡尔曼滤波器能够适应目标的变化行为。
